 Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

 Load Dataset from UCI ML Repository

```python
url = "/content/winequality-red[1].csv"
df = pd.read_csv(url, sep=';')
```

Display First Few Rows

```python
df.head()
```

 Check Dataset Information

```python
df.info()
```

Show Statistical Summary

```python
df.describe()
```

Histogram of Wine Quality

```python
plt.figure(figsize=(8,5))
plt.hist(df["quality"], bins=6, color='skyblue', edgecolor='black')
plt.xlabel("Wine Quality")
plt.ylabel("Frequency")
plt.title("Distribution of Wine Quality Ratings")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Scatter Plot (Alcohol vs. Quality)

```python
plt.figure(figsize=(8,5))
plt.scatter(df["alcohol"], df["quality"], color='red', alpha=0.5)
plt.xlabel("Alcohol Content")
plt.ylabel("Wine Quality")
plt.title("Wine Quality vs Alcohol Content")
plt.grid(True)
plt.show()
```

Line Plot of pH Values

```python
plt.figure(figsize=(8,5))
plt.plot(df.index, df["pH"], color='green', marker='o', linestyle='dashed', markersize=3)
plt.xlabel("Index")
plt.ylabel("pH Level")
plt.title("pH Variation in Wine Samples")
plt.show()
```
